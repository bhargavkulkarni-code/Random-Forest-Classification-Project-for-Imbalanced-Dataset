# under sampling

```python
ns=NearMiss(sampling_strategy=0.6)
x_train_ns,y_train_ns=ns.fit_resample(x_train,y_train)
print('the number of samples before NearMiss',Counter(y_train))
print('the number of samples after NearMiss',Counter(y_train_ns))
```

# over sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro=RandomOverSampler(sampling_strategy=0.7)
x_train_ns,y_train_ns=ro.fit_resample(x_train,y_train)
print('the number of samples before NearMiss',Counter(y_train))
print('the number of samples after NearMiss',Counter(y_train_ns))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=12)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
fc=RandomForestClassifier()
```

```python
fc.fit(x_train_ns,y_train_ns)
```

```python
y_predicted=fc.predict(x_test)
y_predicted
```

```python
fc.score(x_test,y_test)
```

# In the above code 55,400 is same this is o (no fraud) and 1 (fraud) is increased to 38,780.

# so over sampling given more accuracy compare to under sampling we will go forword with over sampling.
